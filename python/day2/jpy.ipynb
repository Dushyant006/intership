```python
import requests

def weather_data(city):
    url=f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=81c9b5bdfffdea12e9540d06757f7af0&units=metric"
    try :
        responce = requests.get(url)
        responce.raise_for_status()
        data= responce.json()
        print(f"city name:{city}")
        print(data['main']['temp'])
        print("city temp feels like")
        print(data['main']['feels_like'])
        print("city minimum temp")
        print(data['main']['temp_min'])
        print("city maximum temp")
        print(data['main']['temp_max'])
        print("city pressure")
        print(data['main']['pressure'])
        print("city humidity")
        print(data['main']['humidity'])
    except requests.exceptions.ReadTimeout as e:
        print(e)

city= input("enter your city name")
weather_data(city)
```
