```python
def max(a, b, c):
    print("Returns the highest value among three numbers.")
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

max(10, 35, 22)
```

```python
def elements(l):
    print("Takes a list and returns a new list with duplicates removed.")
    return list(set(l))

print(elements([1, 2, 2, 3, 4, 4, 4, 5]))
```

```python
def list(n):
    print("Multiplies all the numbers inside a list together.")
    if not n: 
        return 0    
    result = 1
    for num in n:
        result *= num
    return result

print(list([2, 3, 4])) 
```

```python
def factorial(n):
    print("Calculates the factorial of a non-negative integer.")
    
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5)) 
```

```python
def string(text):
    return text[::-1]

print(string("Python")) 
```

```python
def is_in_range(num, start, end):
    """Checks if a number falls within the inclusive range [start, end]."""
    return start <= num <= end

print(is_in_range(7, 1, 10)) 
print(is_in_range(15, 1, 10)) 
```

```python
def even_no(n):
    even_numbers = [num for num in n if num % 2 == 0]
    print(even_numbers)

# Example:
even_no([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
def is_prime(num):
    """Checks if a number is a prime number."""
    if num <= 1:
        return False
    
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
            
    return True

print(is_prime(11)) 
print(is_prime(4))  
```

```python
def count_case(text):
    """Counts and prints the number of uppercase and lowercase letters."""
    counts = {"upper": 0, "lower": 0}
    
    for char in text:
        if char.isupper():
            counts["upper"] += 1
        elif char.islower():
            counts["lower"] += 1
            
    print(f"Original Text: {text}")
    print(f"Uppercase characters: {counts['upper']}")
    print(f"Lowercase characters: {counts['lower']}")

count_case("Hello WORLD")
```
