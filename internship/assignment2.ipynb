```python
fnumber = float(input("Enter the first number: "))
snumber = float(input("Enter the second number: "))

print("\n Operations: add, subtract, multiply, divide")
operation = input("Which operation would you like to perform? ").strip().lower()

if operation == "add":
    result = fnumber + snumber
    print(f"\nResult: {result}")
    
elif operation == "subtract":
    result = fnumber - snumber
    print(f"\nResult: {result}")
    
elif operation == "multiply":
    result = fnumber * snumber
    print(f"\nResult: {result}")
    
elif operation == "divide":
    if snumber == 0:
        print("\nError: Cannot divide by zero!")
    else:
        result = fnumber / snumber
        print(f"\nResult: {result}")
        
else:
    print("\nError: Invalid operation selected.")
```

```python
n = input("Enter a number to check: ")

reversed_n = n[::-1]

if n == reversed_n:
    print(f"Yes, {n} is a Palindrome Number")
else:
    print(f"No, {n} is not a Palindrome Number")
```
