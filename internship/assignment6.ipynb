```python
import requests

def weather_data(city):
    url=f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=3f711726b6d85679413b2138dc9210df&units=metric"
    try :
        responce = requests.get(url)
        responce.raise_for_status()
        data= responce.json()
        print(f"city name:{city}")
        print(data['main']['temp'])
        print("city temp feels like")
        print(data['main']['feels_like'])
        print("city minimum temp")
        print(data['main']['temp_min'])
        print("city maximum temp")
        print(data['main']['temp_max'])
        print("city pressure")
        print(data['main']['pressure'])
        print("city humidity")
        print(data['main']['humidity'])
    except requests.exceptions.ReadTimeout as e:
        print(e)

city= input("enter your city name")
weather_data(city)
```

```python
import random

def play_game():
    choices = ["rock", "paper", "scissors"]
    user_score = 0
    computer_score = 0
    
    print("=== Rock, Paper, Scissors ===")
    print("Type 'quit' to end the game.\n")
    
    while True:
        user_choice = input("Enter Rock, Paper, or Scissors: ").lower().strip()
        
        if user_choice == "quit":
            print("\nThanks for playing!")
            print(f"Final Score -> You: {user_score} | Computer: {computer_score}")
            break
            
        if user_choice not in choices:
            print("Invalid choice. Please try again.\n")
            continue
            
        computer_choice = random.choice(choices)
        print(f"Computer chose: {computer_choice.capitalize()}")
        
        # Determine the winner
        if user_choice == computer_choice:
            print("It's a tie!")
        elif (user_choice == "rock" and computer_choice == "scissors") or \
             (user_choice == "paper" and computer_choice == "rock") or \
             (user_choice == "scissors" and computer_choice == "paper"):
            print("You win this round! 🎉")
            user_score += 1
        else:
            print("Computer wins this round! 🤖")
            computer_score += 1
            
        print(f"Score -> You: {user_score} | Computer: {computer_score}\n")

if __name__ == "__main__":
    play_game()
```

```python
import requests

def get_joke():
    url = "https://official-joke-api.appspot.com/jokes/random"
    response = requests.get(url)
    if response.status_code == 200:
        joke = response.json()
        print("Joke:", joke["setup"])
        print(" punchline:", joke["punchline"])
    else:
        print("Failed to fetch joke")

get_joke()
```
