```python
#Create a Pandas Series from Dictionary 
import pandas as pd

data = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
series = pd.Series(data)
print(series)
print("-"*30)
#Create a Pandas Series from List
data = [1, 2, 3, 4, 5]
series= pd.Series(data)
print(series)
print("-"*30)

#Access the elements of a Series in Pandas
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(series[0])  
print(series[4])
```

```python
#Make a Pandas DataFrame with a two-dimensional Python list
data = [[1, 'Alice'], [2, 'Bob'], [3, 'Charlie']]
columns_name = ['ID', 'Name']
df = pd.DataFrame(data, columns=columns_name)
print(df)
print("-"*30)

#Create DataFrame from Python dict
data = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
df = pd.DataFrame(data)
print(df)
print("-"*30)

#Create Pandas dataframe using list of lists
data = [[1, 'Alice'], [2, 'Bob'], [3, 'Charlie']]
df = pd.DataFrame(data, columns=columns_name)
print(df)
print("-"*30)
#Create a Pandas dataframe using list of tuples
data = [(1, 'Alice'), (2, 'Bob'), (3, 'Charlie')]
df = pd.DataFrame(data, columns=columns_name)
print(df)
print("-"*30)

#Create a Pandas DataFrame from List of Dicts
data = [{'ID': 1, 'Name': 'Alice'}, {'ID': 2, 'Name': 'Bob'}, {'ID': 3, 'Name': 'Charlie'}]
df = pd.DataFrame(data)
print(df)
```

```python
#Different ways to iterate over rows in Pandas Dataframe
data = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
df = pd.DataFrame(data)

for index, row in df.iterrows():
    print(f"Index: {index}, ID: {row['ID']}, Name: {row['Name']}")

for row in df.itertuples():
    print(f"Index: {row.Index}, ID: {row.ID}, Name: {row.Name}")

def print_row(row):
    print(f"ID: {row['ID']}, Name: {row['Name']}")
df.apply(print_row, axis=1)
print("-"*30)


#Selecting rows in pandas DataFrame based on conditions
data = {'ID': [1, 2, 3, 4, 5], 'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve']}
df = pd.DataFrame(data)

new_df = df[df['ID'] > 2]
print(new_df)

new_df = df[df['Name'] == 'Alice']
print(new_df)
print("-"*30)


#Select any row from a Dataframe using iloc[]
data = {'ID': [1, 2, 3, 4, 5], 'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve']}
df = pd.DataFrame(data)
row = df.iloc[0]  
print(row)
print("-"*30)


#Limited rows selection with given column
data = {'ID': [1, 2, 3, 4, 5], 'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve']}
df = pd.DataFrame(data)
selected_rows = df.loc[df['ID'] > 2, ['ID', 'Name']]
print(selected_rows)
print("-"*30)


#Drop rows from the dataframe based on certain condition applied on a column
data = {'ID': [1, 2, 3, 4, 5], 'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve']}
df = pd.DataFrame(data)
updated_df = df[df['ID'] <= 2]
print(updated_df)
print("-"*30)


#Insert row at given position in Pandas Dataframe
data = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
df = pd.DataFrame(data)
new_row = {'ID': 4, 'Name': 'David'}
df = pd.concat([df.iloc[:2], pd.DataFrame([new_row]), df.iloc[2:]]).reset_index(drop=True)
print(df)
print("-"*30)


#Create a list from rows in Pandas dataframe
data = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
df = pd.DataFrame(data)
row_list = df.values.tolist()
print(row_list)
```
