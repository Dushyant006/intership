```python
#Convert a series of date-strings to a timeseries?
import pandas as pd
data = {
    'date_string': ['2026-01-01', '2026-01-02', '2026-01-03'],
    'value': [10, 15, 20]
}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date_string'])
df.set_index('date', inplace=True)
df.drop(columns=['date_string'], inplace=True)
print(df)
```

```python
#Create two DataFrames, df1 and df2, with a common column (e.g., 'ID').
import pandas as pd
data1 = {
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
}
data2 = {
    'ID': [2, 3, 4],
    'Age': [25, 30, 35]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

#Perform an inner merge on this common column and display the resulting DataFrame.
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print(merged_df)
print("-"*30)

left_join_df = pd.merge(df1, df2, on='ID', how='left')
print(left_join_df)

right_join_df = pd.merge(df1, df2, on='ID', how='right')
print(right_join_df)

index_join_df = df1.set_index('ID').join(df2.set_index('ID'), how='inner')
print(index_join_df)
print("-"*30)

#Perform a right join using pd.merge() on a common column, then perform a join using df.join() based on the index. Compare the results. Merging with Multiple Keys
data3 = {
    'ID': [1, 2, 3],
    'Category': ['A', 'B', 'C'],
    'Value': [100, 200, 300]
}
data4 = {
    'ID': [2, 3, 4],
    'Category': ['B', 'C', 'D'],
    'Value': [150, 250, 350]
}
df3 = pd.DataFrame(data3)
df4 = pd.DataFrame(data4)
merged_multi_key = pd.merge(df3, df4, on=['ID', 'Category'], how='right', suffixes=('_df3', '_df4'))
print(merged_multi_key)
index_join_multi_key = df3.set_index(['ID', 'Category']).join(df4.set_index(['ID', 'Category']), how='right', lsuffix='_df3', rsuffix='_df4')
print(index_join_multi_key)


```

```python
#Create three DataFrames. Vertically concatenate two of them using pd.concat(), then merge the resulting DataFrame with the third DataFrame on a common key. T Understand join() vs. merge().
data1 = {
    'ID': [1, 2],
    'Name': ['Alice', 'Bob']
}
data2 = {
    'ID': [3, 4],
    'Name': ['Charlie', 'David']
}
data3 = {
    'ID': [1, 2, 3, 4],
    'Age': [25, 30, 35, 40]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
concatenated_df = pd.concat([df1, df2], ignore_index=True)
merged_df = pd.merge(concatenated_df, df3, on='ID', how='inner')
print(merged_df)
joined_df = concatenated_df.set_index('ID').join(df3.set_index('ID'), how='inner')
print(joined_df)
```
